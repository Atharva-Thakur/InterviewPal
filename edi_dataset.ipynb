```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
train = pd.DataFrame()
prompt = "Below is an question asked to the candidate during an interview for a business startup along with the preferred answer to the question"
for i in range(len(df)):
    train.loc[i, "text"] = prompt + "\n\n### Question :\n" + str(df.iloc[i, 0]) + "\n\n### Answer:\n" + str(df.iloc[i, 1])
```

```python
train
```

```python
train.to_csv("train.csv", index=False)
```
